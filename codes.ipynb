```python
import string
from_ = list(string.ascii_lowercase)

def shift_by(letter,by):

    if letter in from_[0:len(from_) - by]:
        return from_[from_.index(letter) + by]
    elif letter in from_[len(from_) - by : len(from_)]:
        return from_[len(from_) - from_.index(letter)]
    else: 
        return letter
    
#string = "g fmnc wms bgblr rpylqjyrc gr zw fylb. rfyrq ufyr amknsrcpq ypc dmp. bmgle gr gl zw fylb gq glcddgagclr ylb rfyr'q ufw rfgq rcvr gq qm jmle. sqgle qrpgle.kyicrpylq() gq pcamkkclbcb. lmu ynnjw ml rfc spj."
str= "map"
input = list(str)
output = [shift_by(char,2) for char in input]
"".join(output)
```

```python
f = open("txt.txt","r")
input_ = list(f.readlines())
f.close()
input_ = "".join(input_)

def find_rare_char(input_):
    output_ = []
    for i in input_:
        if i in list(string.ascii_lowercase) + list(string.ascii_uppercase):
            output_.append(i)
    return output_

find_rare_char(input_)
```

## Q10

```python
def lookandsay(number):
    result = ""
    repeat = number[0]
    number = number[1:]+" "
    times = 1
    for actual in number:
        if actual != repeat:
            result += str(times)+repeat
            times = 1
            repeat = actual
        else:
            times += 1
    return result
 
num = "1"
 
for i in range(30):
    print (num)
    num = lookandsay(num)
```
